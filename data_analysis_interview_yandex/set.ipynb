```python
from typing import Any, List
class Set:
    def __init__(self) -> None:
        self.divisor = 128
        self.storage:List[List] = [[] for i in range(self.divisor)]
        self.critical = int(self.divisor * 0.1)
        self.size = 0
        self.load_factor = 0.75
    
    def add(
        self,
        elem:int
    )->None:
        index = elem % self.divisor
        if elem not in self.storage[index]:
            self.size += 1
            self.storage[index].append(elem)
            if self.size/self.divisor > self.load_factor:
                self.rehash()

    def check(
        self,
        elem:int
    )->bool:
        index = elem % self.divisor
        if elem not in self.storage[index]:
            return False
        return True

    def delete(
        self,
        elem:int
    )->Any|None:
        index = elem % self.divisor
        if elem not in self.storage[index]:
            return None
        self.storage[index].remove(elem)
        return True

    def rehash(
        self
    ):
        numbers = []
        for bucket in self.storage:
            for el in bucket:
                numbers.append(el)
        self.divisor *= 2
        self.storage:List[List] = [[] for i in range(self.divisor)]
        for el in numbers:
            self.add(el)
```
